```python
outliers=[]
def detect_outlier(Df):
    
    threshold=3
    mean_1 = np.mean(Df)
    std_1 =np.std(Df)
    
    
    for y in Df:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers
```
